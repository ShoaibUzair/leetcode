```python
#Kadane's Algorithm for Maximum Sum Subarray

def maxSumSubArray(nums):
    max_sum = curr_sum = 0
    
    for num in nums:
        curr_sum = max(num,curr_sum+num)
        if curr_sum > max_sum:
            max_sum = curr_sum
    return max_sum

print(maxSumSubArray([-2,1,-3,4,-1,2,1,-5,4]))
print(maxSumSubArray([5,4,-1,7,8]))
print(maxSumSubArray([1]))
```

```python
#Binary Search
#35. Search Insert Position
def searchInsert(nums,target):
    
    left = 0
    right = len(nums)-1
    
    while left <= right:
        mid = (left+right)//2
        
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
            
    return left

print(searchInsert([1,3,5,6],5))
print(searchInsert([1,3,5,6],2))
print(searchInsert([1,3,5,6],7))
print(searchInsert([1,3,5,6],0))
```

```python
#Trees
'''
Leaf Nodes: Nodes with no children. Nodes at the bottom of the tree.
A tree's height is the number of links from its root to the furthest leaf.
Each node in a tree has a depth: the number of links from the root to the node.

Binary Trees
A full binary tree is a binary tree where every node has exactly 0 or 2 children.

A perfect binary tree doesn't have room for any more nodesâ€”unless we increase the tree's height.

A complete binary tree is like a perfect binary tree missing a few nodes in the last level. 
Nodes are filled in from left to right.

Balanced Binary Tree
The heights of its left and right subtrees differ by at most 1.

'''

#Tree Traversals
#DFS
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

#root with its left and right created
root = TreeNode(32)
root.left = TreeNode(22)
root.right = TreeNode(44)

#Completing the root.left Node
root2 = root.left
root2.left = TreeNode(11)
root2.right = TreeNode(24)

#Completing the root.right Node
root3 = root.right
root3.left = TreeNode(42)
root3.right = TreeNode(45)

#Insertion in a Tree
def insert(root,value):
    
    if root == None:
        root = TreeNode(value)
        return root
    
    if value < root.val:
        root.left = insert(root.left,value)
    if value > root.val:
        root.right = insert(root.right,value)
    return root

'''
    Deletion of a node in a TREE
    case 1 : Node has no children. Its a leaf node.
    case 2: Node has a single child.
    case 3 : Node has two children.
    case 1 : Just remove the link from the parent to the child. Parent node
    will now point to null/None.
    case 2: Replace the node with its only child.
    case 3: Replace with the inorder successor so as to maintain the
    inorder traversal of the tree.
'''
    
def minValueNode(root):
    curr = root
    
    while curr.left:
        curr = curr.left
    
    return curr

def maxValueNode(root):
    curr = root
    
    while curr.right:
        curr = curr.right
    
    return curr

def delete(root,value):
    
    if root is None:
        return None
    #Go the right subtree, if the value to be deleted is greater than root value
    if value > root.val:
        root.right = delete(root.right,value)
    #Go the left subtree, if the value to be deleted is lesser than root value
    elif value < root.val:
        root.left = delete(root.left,value)
    else:
        #For handling cases with nodes having a single or no child
        if root.left is None:
            temp = root.right
            root = None
            return temp
        elif root.right is None:
            temp = root.left
            root = None
            return temp
            
        # Now solving for Node with two children.Get the inorder successor
        temp = minValueNode(root.right)
        # Copy the inorder successor's. Content to this node
        root.val = temp.val
        #Delete the inorder successor
        root.right = delete(root.right,value)
    
    return root

#Inorder Traversal
lis = []
def inorder(root):
    if root:
        inorder(root.left)
        lis.append(root.val)
        inorder(root.right)

insert(root,99)
insert(root,1)
delete(root,1)
inorder(root)
print("Inorder Traversal:")
print(lis)
print("Max value:",maxValueNode(root).val)

#Pre-Order Traversal
lis = []
def preorder(root):
    if root:
        lis.append(root.val)
        preorder(root.left)
        preorder(root.right)
        
preorder(root)
print("Preorder Traversal:")
print(lis) 

#Post-Order Traversal
lis = []
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        lis.append(root.val)
        
postorder(root)
print("Postorder Traversal:")
print(lis) 

#Level Order Traversal
queue = [root]

while queue:
    level = []
    for i in range(len(queue)):
        node = queue.pop(0)
        level.append(node.val)
        
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
    print(level)
    
#DFS without Recursion
#Preorder
stack = [root]
res = []
while stack:
    node = stack.pop()
    res.append(node.val)
   
    if node.right:
        stack.append(node.right)
    if node.left:
        stack.append(node.left)
    
print("DFS without Recursion - PreOrder")
print(res)

#DFS without Recursion
#Inorder
stack = []
current = root
res = []
while True:
    
    if current:
        stack.append(current)
        current = current.left
    elif stack:
        current = stack.pop()
        res.append(current.val)
        current = current.right
    else:
        break
        
        
        
print("DFS without Recursion - Inorder")
print(res)
```
